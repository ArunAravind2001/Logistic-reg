```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, StratifiedKFold

model = LogisticRegression()

# Define compatible combinations
param_grid = [
    {'penalty': ['l1'], 'C': [100, 10, 0.1, 0.001], 'solver': ['liblinear', 'saga']},
    {'penalty': ['l2'], 'C': [100, 10, 0.1, 0.001], 'solver': ['liblinear', 'saga', 'newton-cg', 'lbfgs', 'sag']},
    {'penalty': ['elasticnet'], 'C': [100, 10, 0.1, 0.001], 'solver': ['saga'], 'l1_ratio': [0.5]},
    {'penalty': ['none'], 'solver': ['lbfgs', 'newton-cg', 'sag', 'saga']}
]

cv = StratifiedKFold()

grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
grid.fit(x_train, y_train)
```
